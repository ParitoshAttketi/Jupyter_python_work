```python
# Q23 Take 4 student names and their marks as input from the user.
# Store them in a dictionary (name as key, marks as value). Then find and
# print the student with the highest marks.
student = {}

for i in range(4):
    
    name = input("Enter the name  :")
    marks = int(input("Enter the  marks :")) 

    student[name] = marks



highest_mark = max(student,key = student.get)

hight_std_marks =student[highest_mark]

print(f"Highest student name {highest_mark} and Marks {hight_std_marks}")
```
