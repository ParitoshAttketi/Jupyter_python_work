```python
# 11. Reverse a number using while loop.
# Explanation: Use modulo and division to reverse digits.
# Input: 1234
# Output: 4321

num = int(input("enter a number"))

rev=0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num//10
print(rev)
```

```python
# 12. Check if a number is prime using while loop.
# Explanation: Check divisibility from 2 to sqrt(n).
# Input: 17
# Output: Prime Number

num = int(input("Enter a number: "))

if num < 2:
    print("Not a Prime Number")
else:
    i = 2

    while i * i <= num:
        if num % i == 0:
            print("Not a Prime Number")
            break
        i += 1
    else:
        print("Prime Number")
```

```python
# 13 Print Fibonacci series up to n terms using while loop.
# Explanation: Use two variables a=0, b=1 and generate next terms.
# Input: 8
# Output: 0 1 1 2 3 5 8 13

n = int(input("Enter the number of terms: "))

a = 0
b = 1
count = 1

while count <= n:
    print(a, end=" ")
    c = a + b
    a = b
    b = c
    count = count + 1
```

```python
# 16 Find sum of digits of a number using while loop.
# Explanation: Use modulo 10 and integer division.
# Input: 12345
# Output: Sum of digits = 15

n = int(input("Enter the number of terms: "))
rev = 0
while n > 0:
    digit = n % 10
    rev = rev + digit
    n = n // 10
rev
    
```

```python
# 19 Print prime numbers between 1 to 100 using nested for loop.
# Explanation: Outer loop for numbers, inner loop to check divisibility.
# Input: None
# Output: List of primes

for num in range(2, 101):
    count = 0

    for i in range(1, num + 1):
        if num % i == 0:
            count = count + 1

    if count == 2:
        print(num, end=" ")
    
    
```

```python
# 22 Print a hollow square pattern of size n using nested loop.
# Input: 5
# Output: Square with only border *

n = int(input("Enter size: "))

for i in range(n):
    for j in range(n):
        if i == 0 or i == n - 1 or j == 0 or j == n - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```
