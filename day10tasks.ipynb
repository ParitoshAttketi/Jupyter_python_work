```python
# 16. Use while loop to reverse a number (e.g., 1234 → 4321).
num=1234
rev=0
while num >0:
    digit = num % 10
    rev= rev * 10 + digit
    num = num // 10
print(rev)
```

```python
# 18. Print the Fibonacci series up to 10 terms using while loop.
a = 0
b = 1
count = 1

while count <= 10:
    print(a, end=" ")
    c = a + b
    a = b
    b = c
    count = count + 1
```

```python
# 19. Use while loop to check if a given number is prime or not.

num = 1
while num <= 10:
    if num > 1:
        for i in range(2,num):
            if num % i == 0:
                print(f"{num} is not prime number")
                break
        else:
            print(f"{num} is a prime number")    
    num += 1    
```

```python
# 20. Create a simple menu-driven program using while loop (options: 1. Add, 2. Subtract, 3. Exit).

while True:
    print("===========================")
    print("options :")
    print("1 : Add")
    print("2 : Subtract")
    print("3 : exit")
    print("===========================")

    select = int(input("enter ur option"))

    if select == 1:
        print("===========================")
        num1= int(input("enter ur first number"))
        num2= int(input("enter ur second number"))
        print("sum :",num1 + num2)
        print("===========================")
    elif select == 2:
        print("===========================")
        num1= int(input("enter ur first number"))
        num2= int(input("enter ur second number"))
        print("difference :",num1 - num2)
        print("===========================")
    elif select == 3:
        print("===========================")
        print("exit")
        break
        print("===========================")
    else:
        print("invalid input")
        print("===========================")
```
