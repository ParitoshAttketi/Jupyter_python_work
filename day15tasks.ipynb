```python
# 5. Bank Account
# Attributes
# account_holder
# balance
# Methods
# deposit()
# withdraw()
# Withdraw should happen only if balance is sufficient.

class BankAccount:
    def __init__(self,account_holder,balance):

        self.account_holder = account_holder
        self.balance = balance

    def deposit(self,amount):

        self.balance += amount
        print(f"{amount} deposit successfully ")
        print(f"updated balance is {self.balance}")

    def withdraw(self,amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"{amount} withdrawn successfully")
            print(f"remaining balance is {self.balance}")
        else:
            print("Insufficient balance! Withdrawal failed.")

account = BankAccount("paritosh",1000000)

account.deposit(500000)

account.withdraw(1000000)

account.withdraw(500000)

account.withdraw(1)
```

```python
# 9. Library
# Attributes
# book_name
# available_copies
# Method
# borrow_book()
# Reduce copy only if available.

class Library:
    def __init__(self,book_name,available_copies):
        self.book_name = book_name
        self.available_copies = available_copies

    def borrow_book(self):
        if self.available_copies > 0:
            self.available_copies -= 1
            print("book has been borrowed successfully")
            print(f"{self.available_copies} available copies")
            
library = Library("song of ice and fire",5)

library.borrow_book()
```

```python
# 19. ATM
# Attributes
# balance
# Methods
# deposit()
# withdraw()
# withdraw()
# should reject
# negative amount
# zero
# insufficient balance

class ATM:
    def __init__(self, balance):
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"₹{amount} deposited successfully.")
            print(f"Current Balance: ₹{self.balance}")
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount):
        if amount <= 0:
            print("Invalid withdrawal amount.")
        elif amount > self.balance:
            print("Insufficient balance.")
        else:
            self.balance -= amount
            print(f"₹{amount} withdrawn successfully.")
            print(f"Remaining Balance: ₹{self.balance}")

atm = ATM(5000)

atm.deposit(2000)

atm.withdraw(3000)

atm.withdraw(0)
atm.withdraw(-500)

atm.withdraw(5000)
```

```python
# 21. Inventory System
# Attributes
# product_name
# stock
# Method
# update_stock(change)
# Positive value → add stock
# Negative value → remove stock
# Stock should never become negative.

class Inventory:
    def __init__(self, product_name, stock):
        self.product_name = product_name
        self.stock = stock

    def update_stock(self, change):
        if change > 0:
            self.stock += change
            print(f"{change} items added.")
        elif self.stock + change >= 0:
            self.stock += change
            print(f"{-change} items removed.")
        else:
            print("Stock cannot become negative.")

        print(f"Product: {self.product_name}")
        print(f"Current Stock: {self.stock}")

item = Inventory("Laptop", 20)

item.update_stock(10)

item.update_stock(-15)

item.update_stock(-20)
```

```python
# 23. Student Attendance
# Attributes
# total_classes
# attended_classes
# Method
# mark_attendance(number)
# Attendance should never exceed total classes.
# Print attendance percentage after every update.

class StudentAttendance:
    def __init__(self, total_classes, attended_classes):
        self.total_classes = total_classes
        self.attended_classes = attended_classes

    def mark_attendance(self, number):
        if self.attended_classes + number <= self.total_classes:
            self.attended_classes += number
            percentage = (self.attended_classes / self.total_classes) * 100

            print(f"Attendance marked for {number} class(es).")
            print(f"Attended Classes: {self.attended_classes}")
            print(f"Attendance Percentage: {percentage:.2f}%")
        else:
            print("Attendance cannot exceed total classes.")

student = StudentAttendance(100, 75)

student.mark_attendance(10)

student.mark_attendance(20)
```

```python
# 24. Bus Reservation (Hard)
# Attributes
# bus_name
# total_seats
# booked_seats
# Method
# reserve(number)
# Reservation allowed only if
# booked_seats + number
# does not exceed total seats.
# After booking print
# Remaining seats.

class BusReservation:
    def __init__(self, bus_name, total_seats, booked_seats):
        self.bus_name = bus_name
        self.total_seats = total_seats
        self.booked_seats = booked_seats

    def reserve(self, number):
        if self.booked_seats + number <= self.total_seats:
            self.booked_seats += number
            remaining_seats = self.total_seats - self.booked_seats

            print(f"{number} seat(s) reserved successfully.")
            print(f"Booked Seats: {self.booked_seats}")
            print(f"Remaining Seats: {remaining_seats}")
        else:
            print("Reservation failed! Not enough seats available.")

bus = BusReservation("Express Bus", 50, 40)

bus.reserve(5)
bus.reserve(10)
```

```python
# 25. Online Shopping Cart (Interview Level)
# Attributes
# customer_name
# cart_total
# wallet_balance
# Methods
# add_item(price)
# remove_item(price)
# checkout()
#Rules
# Cart total cannot become negative.
# Checkout only if wallet balance is enough.
# Deduct wallet balance after successful checkout.
# Cart becomes zero after payment.
# Print suitable messages for every case.

class ShoppingCart:
    def __init__(self, customer_name, wallet_balance):
        self.customer_name = customer_name
        self.cart_total = 0
        self.wallet_balance = wallet_balance

    def add_item(self, price):
        self.cart_total += price
        print(f"Item of ₹{price} added.")
        print(f"Cart Total: ₹{self.cart_total}")

    def remove_item(self, price):
        if self.cart_total - price >= 0:
            self.cart_total -= price
            print(f"Item of ₹{price} removed.")
            print(f"Cart Total: ₹{self.cart_total}")
        else:
            print("Cart total cannot become negative.")

    def checkout(self):
        if self.cart_total == 0:
            print("Your cart is empty.")
        elif self.wallet_balance >= self.cart_total:
            self.wallet_balance -= self.cart_total
            print("Payment successful.")
            print(f"Amount Paid: ₹{self.cart_total}")
            self.cart_total = 0
            print(f"Wallet Balance: ₹{self.wallet_balance}")
            print(f"Cart Total: ₹{self.cart_total}")
        else:
            print("Insufficient wallet balance.")

cart = ShoppingCart("Rahul", 5000)

cart.add_item(1500)
cart.add_item(1000)

cart.remove_item(500)

cart.checkout()

cart.checkout()
```
