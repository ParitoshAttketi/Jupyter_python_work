```python
# bank managment

class Bank_management:
    def __init__(self,acc_no,bal,branch):
        self.Account_number = acc_no
        self.balance = bal
        self.branch = branch

    def display_bank_balance(self):
        print(f"{self.Account_number} : Account number")
        print(f"{self.branch} : Account branch")
        

    def deposit_money(self,amount):
        self.balance += amount
        print(f"{amount}: deposit money successfully")
        print(f"{self.balance} : total balance")

    def withdraw_money(self,amount):

        if amount > 0:
            if amount <= self.balance:
                self.balance -= amount
                print(f"{amount}: withdraw money successfully")
                print(f"{self.balance} : total balance")
            else:
                print("insuffient balance")


bank = Bank_management(112233,1000000,"ameerpet")

bank.display_bank_balance()

bank.deposit_money(50000)

bank.withdraw_money(10500000)
```

```python

class Animals(ABC):
    # abstractmethod
    @abstractmethod
    def sound(self):
        pass
# child class
class Cat(Animals):
    def sound(self):
        print("meoww")
class Dog(Animals):
    def sound(self):
        print("bhauu")
class cow(Animals):
    def sound(self):
        print("moo")

# class object
cat= Cat()
cat.sound()
```

```python
class Food(ABC):
    def __init__(self,f_name,f_price):
        self.__f_name = f_name
        self.__f_price = f_price

    def get_food_name(self):
        return self.__f_name

    def get_food_price(self):
        return self.__f_price

    @abstractmethod
    def prepare(self):
        pass

    @abstractmethod
    def deliver(self):
        pass

    # display all details
    def display_details(self):
        print(f"food name : {self.__f_name}")
        print(f"food price : {self.__f_price}")
        self.prepare()
        self.deliver()
        
# child class
class Pizza(Food):
    def prepare(self):
        print("pizza is prepared")
    def deliver(self):
        print("pizza is delivered")

class Burger(Food):
    def prepare(self):
        print("burger is prepared")
    def deliver(self):
        print("burger is delivered")
        
class Biriyani(Food):
    def prepare(self):
        print("biriyani is prepared")
    def deliver(self):
        print("biriyani is delivered")    


# create class object
pizza = Pizza("pizza",500)
burger = Burger("burger",300)
biriyani = Biriyani("biriyani",300)

pizza.get_food_name()
pizza.get_food_price()
pizza.display_details()
```

```python
from abc import ABC, abstractmethod

class Payment(ABC):
    
    def __init__(self,owner,amount):
        self.__ownername = owner
        self.__amount = amount

    def get_ownername(self):
        return f"Owner Name : {self.__ownername}"

    def get_amount(self):
        return f"amount : {self.__amount}"

    def set_amount(self,new):
        if new > 0:
            self.__amount += new
            return f"amount updated : {self.__amount}"
        else:
            print("please enter a valid amount")

    @abstractmethod
    def pay(self):
        pass 

    def recipt(self):
        print(f"owner name : {self.__ownername}")
        print(f"amount : {self.__amount}")

class Credit_card(Payment):
    def __init__(self,owner,amount,credit_no):
        super().__init__(owner,amount)                                 #### super()
        self.__Credit_no = credit_no

    def pay(self):
        print("Credit card details :")
        print(f"Credit card number : {self.__Credit_no}")
        self.recipt()

class Debit_card(Payment):
    def __init__(self,owner,amount,dedit_no):
        super().__init__(owner,amount)                                 #### super()
        self.__Dedit_no = dedit_no

    def pay(self):
        print("Dedit card details :")
        print(f"Dedit card number : {self.__Dedit_no}")
        self.recipt()

class UPI(Payment):
    def __init__(self,owner,amount,upi_no):
        super().__init__(owner,amount)                                 #### super()
        self.__Upi_no = upi_no

    def pay(self):
        print("UPI details :")
        print(f"UPI number : {self.__Upi_no}")
        self.recipt()

credit = Credit_card("sumit",1000,2000)
debit = Debit_card("paritosh",1500,2000)
upi = UPI("yesh",10000,12345)

credit.pay()
print("*" * 50)
debit.pay()
print("*" * 50)
upi.pay()
```

```python
class Agent(ABC):
    def __init__(self,name,apikey):
        self.Agent_name = name
        self.__Agent_apikey = apikey

    @abstractmethod
    def LLM_(self):
        pass

    def Agent_info(self):
        print(f"Agent name : {self.Agent_name}")
        print(f"Agent apikey : {self.__Agent_apikey}")

class Flight(Agent):
    def __init__(self,name,apikey,role):
        super().__init__(name,apikey)
        self.Agent_role = role

    def LLM_(self):
        return "Open AI LLM model"


    def display_details(self):
        print(self.LLM_())
        print(self.Agent_info())
        print(f"Agent role : {self.Agent_role}")
        
class Hostel(Agent):
    def __init__(self,name,apikey,role):
        super().__init__(name,apikey)
        self.Agent_role = role

    def LLM_(self):
        return "GROK AI LLM model"
        
    def display_details(self):
        print(self.LLM_())
        print(self.Agent_info())
        print(self.Agent_role())


flight = Flight("paritosh",12345,"flight booking")
hostel = Hostel("sumit",2345,"hostel booking")

flight.display_details()
print("*" * 20)
hostel.display_details()
```
