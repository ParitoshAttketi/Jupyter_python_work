```python
# 6. Class Name: BankAccount → SavingsAccount
# Problem
# Create a parent class named BankAccount.
# Explanation
# Provide methods:
# deposit()
# withdraw()
# Create subclass:
# SavingsAccount
# Add:
# calculate_interest()

class BankAccount:
    def __init__(self,balance):
        self.balance = balance 
        
    def deposit(self,amount):
        self.balance += amount
        print("the amount has been debited")
        print(f"balance : {self.balance}")

    def withdraw(self,amount):
        if amount <= self.balance:
            self.balance -= amount
            print("withdraw has been done successfully")
            print(f"remainint balance : {self.balance}")
        else:
            print("invalid amount")

class SavingsAccount(BankAccount):
    def __init__(self,balance):
        super().__init__(balance)

    def calculate_interest(self):
        interest = self.balance * 0.10
        print(f"Interest value : {interest}")

c = SavingsAccount(25000)
c.deposit(5000)
c.withdraw(10000)
c.calculate_interest()
```

```python
# 12. Class Name: Account → LoanAccount
# Problem
# Create a parent class named Account.
# Explanation
# Store:
# balance
# Create subclass:
# LoanAccount
# Add:
# calculate_emi()

class Account:
    def __init__(self,balance):
        self.balance = balance 

class LoanAccount(Account):
    def __init__(self,balance):
        super().__init__(balance)

    def calculate_emi(self,inst):
        if inst > 0:
            self.inst = inst
            self.balance = self.balance + (self.balance * 0.10)
            print(f"Amount with EMI : {self.balance}")
        else:
            print("invalid input")
            
c = LoanAccount(40000)
c.calculate_emi(5)
```

```python
# 30. Class Name:
# GrandParent → Parent → Child → GrandChild
# Problem
# Create four levels of inheritance.
# Explanation
# Each class should have its own method.
# The GrandChild object should be able to access methods from all parent classes.
# Demonstrate Python's Method Resolution Order (MRO) using:
# ClassName.mro()
# or
# help(ClassName)

class GrandParent:
    def grandparent_method(self):
        print("This is GrandParent class")

class Parent(GrandParent):
    def parent_method(self):
        print("This is Parent class")

class Child(Parent):
    def child_method(self):
        print("This is Child class")

class GrandChild(Child):
    def grandchild_method(self):
        print("This is GrandChild class")

a = GrandChild()

a.grandparent_method()
a.parent_method()
a.child_method()
a.grandchild_method()

print("Method Resolution Order (MRO):")
print(GrandChild.mro())
```
