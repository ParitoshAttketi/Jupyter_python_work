```python
# 16. Create a function calculate(a, b, operation) and show how keyword args improve readability.
def calculate(a, b, operation):
    if operation == "add":
        return a + b
    elif operation == "sub":
        return a - b
    elif operation == "multiply":
        return a * b
    elif operation == "divide":
        if b != 0:
            return a / b
        else:
            return "cant divide by 0"       
    else:
        return "invalid operation"
```

```python
calculate(a=10 , b=20, operation ="add")
```

```python
# Write a function process_data(*args, **kwargs) that:
# Takes any number of positional arguments (numbers)
# Takes keyword arguments where keys are operation names ("add", "multiply","power")
# Performs the requested operations on all numbers
# Returns a dictionary with results Example: process_data(2, 3, 4, add=True, multiply=True) should return appropriate results.

def process_data(*args, **kwargs):
    result = {}

    if kwargs.get("add"):
        result["add"] = sum(args)

    if kwargs.get("multiply"):
        product = 1
        for num in args:
            product *= num
        result["multiply"] = product

    if kwargs.get("power"):
        power_list = []
        for num in args:
            power_list.append(num ** 2)   # Square each number
        result["power"] = power_list

    return result
```

```python
process_data(2, 3, 4, add=True, multiply=True, power=True)
```
