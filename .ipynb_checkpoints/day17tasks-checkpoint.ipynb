```python
# 16. Class Name: Shape
# Problem
# Create an abstract class Shape.
# Explanation
# Declare abstract methods
# calculate_area()
# calculate_perimeter()
# Create subclasses like
# Circle
# Rectangle
# Implement both methods.

from abc import ABC, abstractmethod

import math

class Shape(ABC):

    @abstractmethod
    def calculate_area(self):
        pass

    @abstractmethod
    def calculate_perimeter(self):
        pass

class Circle(Shape):

    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return math.pi * self.radius ** 2

    def calculate_perimeter(self):
        return 2 * math.pi * self.radius

class Rectangle(Shape):

    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculate_area(self):
        return self.length * self.width

    def calculate_perimeter(self):
        return 2 * (self.length + self.width)

circle = Circle(7)
rectangle = Rectangle(10, 5)

print("Area:", circle.calculate_area())
print("Perimeter:", circle.calculate_perimeter())

print("Area:", rectangle.calculate_area())
print("Perimeter:", rectangle.calculate_perimeter())
```

```python
# 21. Class Name: BankAccount
# Problem
# Create abstract class BankAccount.
# Explanation
# Abstract method
# calculate_interest()
# Implement in
# SavingsAccount
# CurrentAccount
# Each account type should use a different interest calculation.

class BankAccount(ABC):
    def __init__(self,balance):
        self.balance = balance

    @abstractmethod
    def calculate_interest(self):
        pass

class SavingsAccount(BankAccount):
    def calculate_interest(self):
        interest = self.balance * 0.05
        print(f"Savings Account interest :{interest}")
        return interest


class CurrentAccount(BankAccount):
    def calculate_interest(self):
        interest = self.balance * 0.02
        print(f"Current Account interest :{interest}")
        return interest
        
bnk = SavingsAccount(1000000)
cnk = CurrentAccount(1000000)

bnk.calculate_interest()
cnk.calculate_interest()
```

```python
# Problem
# Create abstract class Calculator.
# Explanation
# Abstract method
# compute(a, b)
# Implement in
# AddCalculator
# MultiplyCalculator
# Each subclass performs a different operation.

class Calculator(ABC):

    @abstractmethod
    def compute(self, a, b):
        pass

class AddCalculator(Calculator):

    def compute(self, a, b):
        return a + b

class MultiplyCalculator(Calculator):

    def compute(self, a, b):
        return a * b

add = AddCalculator()
multiply = MultiplyCalculator()

print("Addition:", add.compute(10, 5))
print("Multiplication:", multiply.compute(10, 5))
```

```python
# 24. Class Name: Employee
# Problem
# Create abstract class Employee.
# Explanation
# Abstract method
# calculate_bonus()
# Create subclasses
# Manager
# Developer
# HR
# Each has its own bonus calculation rule.

class Employee(ABC):

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    @abstractmethod
    def calculate_bonus(self):
        pass

class Manager(Employee):

    def calculate_bonus(self):
        bonus = self.salary * 0.20      
        print(f"Manager Bonus: {bonus}")
        return bonus

class Developer(Employee):

    def calculate_bonus(self):
        bonus = self.salary * 0.15      
        print(f"Developer Bonus: {bonus}")
        return bonus

class HR(Employee):

    def calculate_bonus(self):
        bonus = self.salary * 0.10 
        print(f"HR Bonus: {bonus}")
        return bonus

manager = Manager("Rahul", 100000)
developer = Developer("Priya", 80000)
hr = HR("Amit", 60000)

manager.calculate_bonus()
developer.calculate_bonus()
hr.calculate_bonus()
```

```python
# 25. Class Name: Shape3D
# Problem
# Create abstract class Shape3D.
# Explanation
# Abstract methods
# volume()
# surface_area()
# Implement in
# Cube
# Sphere
# Cylinder
# Each uses different formulas.

class Shape3D(ABC):

    @abstractmethod
    def volume(self):
        pass

    @abstractmethod
    def surface_area(self):
        pass
  
class Cube(Shape3D):
    def __init__(self, side):
        self.side = side

    def volume(self):
        return self.side ** 3

    def surface_area(self):
        return 6 * self.side ** 2

class Sphere(Shape3D):

    def __init__(self, radius):
        self.radius = radius

    def volume(self):
        return (4 / 3) * math.pi * self.radius ** 3

    def surface_area(self):
        return 4 * math.pi * self.radius ** 2

class Cylinder(Shape3D):

    def __init__(self, radius, height):
        self.radius = radius
        self.height = height

    def volume(self):
        return math.pi * self.radius ** 2 * self.height

    def surface_area(self):
        return 2 * math.pi * self.radius * (self.radius + self.height)

cube = Cube(4)
sphere = Sphere(3)
cylinder = Cylinder(3, 5)

print("Volume:", cube.volume())
print("Surface Area:", cube.surface_area())

print("Volume:", sphere.volume())
print("Surface Area:", sphere.surface_area())

print("Volume:", cylinder.volume())
print("Surface Area:", cylinder.surface_area())
```

```python
# 29. Class Name: Transport
# Problem
# Create abstract class Transport.
# Explanation
# Abstract methods
# move()
# fuel_type()
# Implement in
# Car
# Train
# Airplane
# Each should define:
# how it moves
# what fuel or power source it uses (e.g., Petrol, Diesel, Electricity, Aviation Fuel).

class Transport(ABC):

    @abstractmethod
    def move(self):
        pass

    @abstractmethod
    def fuel_type(self):
        pass

class Car(Transport):

    def move(self):
        print("Car moves on roads.")

    def fuel_type(self):
        print("Fuel Type: Petrol")

class Train(Transport):

    def move(self):
        print("Train moves on railway tracks.")

    def fuel_type(self):
        print("Fuel Type: Electricity")

class Airplane(Transport):

    def move(self):
        print("Airplane flies in the sky.")

    def fuel_type(self):
        print("Fuel Type: Aviation Fuel")

car = Car()
train = Train()
airplane = Airplane()

car.move()
car.fuel_type()

train.move()
train.fuel_type()

airplane.move()
airplane.fuel_type()
```

```python
# 30. Class Name: FileHandler
# Problem
# Create abstract class FileHandler.
# Explanation
# Abstract methods
# read()
# write(data)
# Implement in
# TextFileHandler
# CSVFileHandler
# Each subclass should define how it reads from and writes to its respective file type.

class FileHandler(ABC):

    @abstractmethod
    def read(self):
        pass

    @abstractmethod
    def write(self, data):
        pass

class TextFileHandler(FileHandler):

    def read(self):
        print("Reading data from a text file.")

    def write(self, data):
        print(f"Writing '{data}' to a text file.")

class CSVFileHandler(FileHandler):

    def read(self):
        print("Reading data from a CSV file.")

    def write(self, data):
        print(f"Writing '{data}' to a CSV file.")

text_file = TextFileHandler()
csv_file = CSVFileHandler()

text_file.read()
text_file.write("Hello, World!")

csv_file.read()
csv_file.write("paritosh,24")
```
