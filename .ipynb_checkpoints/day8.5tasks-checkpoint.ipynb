```python
# 06 Write a program to check whether a year is a leap year or not.
year=2024
if year % 4 and (year % 100 != 0) or year % 400:
    print(f"{year} is a leap year")
else:
    print(f"{year} is not leap year")
```

```python
# 10 Write a program to calculate electricity bill:
# Units < 100 → ₹2/unit
# 100–300 → ₹4/unit
# Above 300 → ₹6/unit
units = 250
if units < 100:
    print("₹2/unit")
elif 100 < units < 300:
    print("₹4/unit")
else:
    print("₹6/unit")
```

```python
# 11 Write a program to check whether a triangle is:
# Equilateral
# Isosceles
# Scalene
ang1=int(input(""))
ang2=int(input(""))
ang3=int(input(""))
if (ang1 + ang2) > ang3 and (ang2 + ang3) > ang1 and (ang1 + ang3) > ang2:
    print("valid triangle")
    if ang1 == ang2 == ang3:
        print("Equilateral Triangle")
    elif ang1 == ang2 or ang2 == ang3 or ang1 == ang3:
        print("Isosceles Triangle")
    else:
        print("Scalene Triangle")
else:
    print("not a valid triangle")
```

```python
# 14 Write a program to check whether a number is:
# Multiple of both 3 and 5
# Multiple of only 3
# Multiple of only 5
# None
numb=15
if (numb % 3 == 0) and (numb % 5 == 0):
    print("FizzBuzz")
elif numb % 3 == 0:
    print("Fizz")
elif numb % 5 == 0:
    print("Buzz")
else:
    print(numb)
```

```python
# 18 Write a program to calculate income tax:
# Income below ₹250000 → No tax
# ₹250000–₹500000 → 5%
# ₹500000–₹1000000 → 20%
# Above ₹1000000 → 30%
income = float(input("Enter your annual income: ₹"))

if income < 250000:
    tax = 0
elif income <= 500000:
    tax = income * 0.05
elif income <= 1000000:
    tax = income * 0.20
else:
    tax = income * 0.30

print("Income: ₹", income)
print("Income Tax: ₹", tax)
```
