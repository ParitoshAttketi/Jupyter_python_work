```python
# 1. Take a number as input. If the number is positive, check whether it is even or odd and print the result.
numb=int(input("enter a number"))
if numb > 0:
    print("number is positive")
    if numb % 2 == 0:
        print("number is even")
    else:
        print("number is odd")
else:
    print("number is negative")
```

```python
# 3. Take a single character. If it is an alphabet, check whether it is uppercase or lowercase.
char = "a"
if char.isalpha():
    print("character is alphabet")
    if char.isupper():
        print("character is in uppercase")
    else:
        print("character is in lowercase")
else:
    print("character is not alphabet")
```

```python
# 4. Take three numbers as input and print the largest number using nested if statements (without using max()).
num1=int(input("enter ur number"))
num2=int(input("enter ur number"))
num3=int(input("enter ur number"))

if num1 > num2:
    if num1 > num3:
        print(f"{num1} is larger")
    else:
        print(f"{num3} is larger")
else:
    if num2 > num3:
        print(f"{num2} is larger")
    else:
        print(f"{num3} is larger")
```

```python
# 8. Take three sides of a triangle. First check if they form a valid triangle, then check if it is Equilateral, Isosceles, or Scalene.
side1= 20
side2=20
side3=20

if (side1 + side2) > side3 and (side2 + side3) > side1 and (side3 + side1) > side1:
    print("valid triangle")
    if side1 == side2 == side3:
        print("Equilateral triangle")
    elif side1 == side2 or side2 == side3 or side1 == side3:
        print("Isosceles triangle")
    else:
        print("Scalene")
else:
    print("invalid triangle")
```

```python
# 10.Take a character. If it is a letter, check if it is vowel or consonant. If it is a digit,
# print "Digit". Else print "Special Character".
char = "a"
if char.isalpha():
    print("character is a alphabit")
    if char == "a" or char == "e" or char == "i" or char == "o" or char == "u":
        print("character is vowel")
    else:
        print("character is consonant")
elif char.isdigit():
    print("character is digit")
else:
    print("character is special character")
```

```python
# 11.Take a year as input. Check if it is a leap year using nested conditions.
year = int(input("enter the year"))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("year is a leap year")
        else:
            print("not a leap year")
    else:
        print("leap year")
else:
    print("not a leap year")
```

```python
# 18.Take three sides of a triangle. Check if valid, then check if it is a right-angled
# triangle using Pythagoras theorem.
side1=int(input("enter the val"))
side2=int(input("enter the val"))
side3=int(input("enter the val"))
if (side1+side2)>side3 or (side2+side3)>side1 or (side3+side1)>side2:
     if (side1 ** 2 + side2 ** 2 ) == side3 ** 2 or (side2 ** 2 +side3 ** 2 ) == side1 **2 or (side3 ** 2 + side1 ** 2) == side2:
         print("right-angled triangle")
     else:
         print("not right-angled triangle triangle")
else:
    print("invalid triangle")
    
```

```python
# 20.Take income. If income > 1000000, check tax slab. If income between 500000-
# 1000000, different slab. Else print "Low Income".
income= int(input("enter ur income"))

if income > 1000000:
    print("High Income")
    if income > 2000000:
        print("Tax Slab: 35%")
    else:
        print("Tax Slab: 30%")

elif income >= 500000:
    print("Middle Income")
    if income >= 750000:
        print("Tax Slab: 20%")
    else:
        print("Tax Slab: 10%")

elif income >= 250000:
    print("Lower Middle Income")
    print("Tax Slab: 5%")

else:
    print("Low Income")
    print("No Tax")
```

```python
# 21. (Hard) Take three sides of a triangle as input. First check if they can form a valid
# triangle (sum of any two sides > third side). If valid, then further check and print
# whether the triangle is Equilateral, Isosceles, or Scalene.

ag1=int(input("enter val:"))
ag2=int(input("enter val:"))
ag3=int(input("enter val:"))

if (ag1+ag2)>ag3 or (ag2+ag3)>ag1 or (ag1+ag3)>ag2:
    print("valid triangle")
    if ag1 == ag2 == ag3:
        print("Equilateral triangle")
    elif ag1 == ag2 or ag2 == ag3 or ag1 == ag3:
        print("Isosceles triangle")
    else:
        print("Scalene")
else:
    print("invalid triangle")
```

```python
# 25 Take three numbers. First check if all three are equal. If not, check which
# one is the largest. Also check if the largest number is even or odd and print all
# information accordingly.

num1 = int(input("enter a number"))
num2 = int(input("enter a number"))
num3 = int(input("enter a number"))

if num1 == num2 == num3:
    print("all three values are equal")
else:
    if num1 > num2 and num1 > num3:
        if num1 % 2 == 0:
            print(f"{num1} is larger and even number")
        else:
            print(f"{num1} is larger and odd number")
    elif num2 > num1 and num2 > num3:
        if num2 % 2 == 0:
            print(f"{num2} is larger and even number")
        else:
            print(f"{num2} is larger and odd number")
    else:
        if num3 % 2 == 0:
            print(f"{num3} is larger and even number")
        else:
            print(f"{num3} is larger and odd number")


```
