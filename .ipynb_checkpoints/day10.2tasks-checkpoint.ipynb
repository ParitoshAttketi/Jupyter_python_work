```python
# 1. Print Numbers from 1 to 20
# Write a program to print numbers from 1 to 20 using a while loop.
# Explanation
# Start from 1
# Print the number
# Increase the value by 1 each time
# Stop when number becomes greater than 20
a=1
while a <=20:
    print(a)
    a += 1
```

```python
# 2. Print Even Numbers
# Question
# Write a program to print all even numbers between 1 and 50 using a while loop.
# Explanation
# Even numbers are divisible by 2
# Use loop from 1 to 50
# Check condition using if num % 2 == 0

a=1
while a <= 50:
    if a % 2 == 0:
        print(a)
    a += 1
```

```python
# 3. Sum of Natural Numbers
# Question
# Write a program to calculate the sum of first n natural numbers.
# Explanation

# Example:
# 1 + 2 + 3 + 4 + 5 = 15
# Take value of n
# Use loop from 1 to n
# Add each number into a sum variable

n = int(input("Enter the value of n: "))

i = 1
sum = 0

while i <= n:
    sum = sum + i
    i = i + 1

print("Sum of first", n, "natural numbers is:", sum)
```

```python
# Reverse a Number
# Question
# Write a program to reverse a number using a while loop.
# Explanation
# Example:
# 1234 → 4321
# Steps:
# Take last digit using % 10
# Add digit into reverse number
# Remove last digit using // 10

num = 1234

rev = 0

while num >0:

    digit = num % 10

    rev = rev * 10 + digit

    num = num // 10

print(rev)
```

```python
# 5. Count Digits in a Number
# Question
# Write a program to count total digits in a number.
# Explanation
# Example:
# 56789 → 5 digits
# Steps:
#  Use loop until number becomes 0
#  Remove one digit each time
# Increase counter by 1

num = int(input("Enter a number: "))

count = 0

while num > 0:
    num = num // 10
    count = count + 1

print("Total digits:", count)
```

```python
# 6. Palindrome Number
# Question
# Write a program to check whether a number is palindrome or not.
# Explanation
# Palindrome means same from both sides.
# Example:
# 121 → Palindrome
# 1331 → Palindrome
# Steps:
# Reverse the number
# Compare with original number

num = int(input("enter ur number"))
original = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10
if original == reverse :
    print("it is a palindrome")
else:
    print("it is not a palindrome")
     
```

```python
# 7.Write a program to check whether a number is Armstrong number.
# Explanation
# Example:
# 153
# Calculation:
# 13 + 53 + 33 = 153
# If sum equals original number → Armstrong number

num = int(input("Enter a number: "))

original = num
sum = 0

while num > 0:
    digit = num % 10
    sum = sum + (digit ** 3)
    num = num // 10

if sum == original:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

```python
# 9 Write a program for number guessing game.
# Explanation
# Store one secret number
# Ask user to guess
# If wrong → ask again
# If correct → print success message
secret_number = 7
user_guess = int(input("guess the number"))

while user_guess != secret_number:
    print("wrong, guess again")
    user_guess = int(input("guess the number"))

print("correct guess")
```

```python
# 11 Write a program to print Fibonacci series.
# Explanation
# Series:
# 0 1 1 2 3 5 8 13 ...
# Each next number is sum of previous two numbers.
n = int(input("Enter the number of terms: "))

a = 0
b = 1
count = 1

while count <= n:
    print(a, end=" ")
    c = a + b
    a = b
    b = c
    count = count + 1
```

```python
# 12 Write a program to find factorial of a number.
# Explanation
# Example:
# 5! = 5 × 4 × 3 × 2 × 1 = 120
# Multiply numbers one by one using loop

num = int(input("Enter a number: "))

factorial = 1
i = 1

while i <= num:
    factorial = factorial * i
    i = i + 1

print("Factorial =", factorial)
```

```python
# 13 Create calculator using while loop and if elif else.
# Explanation
# Menu Example:
# 1. Add
# 2. Subtract
# 3. Multiply
# 4. Divide
# 5. Exit

while True:
    print("===========================")
    print("options :")
    print("1 : Add")
    print("2 : Subtract")
    print("3 : Multiply")
    print("4 : Divide")
    print("5 : exit")
    print("===========================")

    select = int(input("enter ur option"))

    if select == 1:
        print("===========================")
        num1= int(input("enter ur first number"))
        num2= int(input("enter ur second number"))
        print("sum :",num1 + num2)
        print("===========================")
    elif select == 2:
        print("===========================")
        num1= int(input("enter ur first number"))
        num2= int(input("enter ur second number"))
        print("difference :",num1 - num2)
        print("===========================")
    elif select == 3:
        print("===========================")
        num1= int(input("enter ur first number"))
        num2= int(input("enter ur second number"))
        print("difference :",num1 * num2)
        print("===========================")
    elif select == 4:
        print("===========================")
        num1= int(input("enter ur first number"))
        num2= int(input("enter ur second number"))
        print("difference :",num1 / num2)
        print("===========================")
    elif select == 5:
        print("============================")
        print("exit")
        print("===========================")
        break
    else:
        print("invalid input")
        print("===========================")
```

```python
# 14 Create simple ATM program.
# Explanation
# Options:
# Check balance
# Deposit money
# Withdraw money
# Exit
# Use loop so menu repeats again and again.
    
balance = 1000

while True:
    print("----- ATM Menu -----")
    print("1. Check Balance")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        print("Balance =", balance)

    elif choice == 2:
        amount = int(input("Enter amount to deposit: "))
        balance = balance + amount
        print("Amount deposited successfully.")
        print("New Balance =", balance)

    elif choice == 3:
        amount = int(input("Enter amount to withdraw: "))
        if amount <= balance:
            balance = balance - amount
            print("Please collect your cash.")
            print("Remaining Balance =", balance)
        else:
            print("Insufficient Balance")

    elif choice == 4:
        print("Thank you for using the ATM.")
        break

    else:
        print("Invalid Choice")
```

```python
# 15 Take numbers from user continuously and stop when user enters 0.
# Explanation
# Add all numbers into sum
# When input becomes 0 → stop loop
# Print final sum
sum = 0

num = int(input("Enter a number (0 to stop): "))

while num != 0:
    sum = sum + num
    num = int(input("Enter a number (0 to stop): "))

print("Final Sum =", sum)
```

```python
# 16 Check leap year repeatedly until user types "stop".
# Explanation
# Ask year from user
# Check leap year condition
# Continue until "stop" entered
# Leap year condition:
# Divisible by 4
# Not divisible by 100 OR divisible by 400

year = input("Enter year (or type 'stop' to exit): ")

while year != "stop":
    year = int(year)

    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        print("Leap Year")
    else:
        print("Not a Leap Year")

    year = input("Enter year (or type 'stop' to exit): ")

print("Program Ended")
```

```python
# 18 Print all prime numbers from 1 to 100.
# Explanation
# Prime number:
# Divisible only by 1 and itself
# Example:
# 2, 3, 5, 7, 11 ...
# Use nested loop to check divisibility.

num = 2

while num <= 100:
    i = 2
    prime = True

    while i < num:
        if num % i == 0:
            prime = False
        i = i + 1

    if prime:
        print(num)

    num = num + 1
```

```python
# 19 Keep taking student marks until negative number entered.
# Explanation
# Add all marks
# Count total students
# Stop on negative input
# Find average:
# average = total / count
total = 0
count = 0

marks = int(input("Enter student marks (negative to stop): "))

while marks >= 0:
    total = total + marks
    count = count + 1
    marks = int(input("Enter student marks (negative to stop): "))

if count > 0:
    average = total / count
    print("Total Marks =", total)
    print("Total Students =", count)
    print("Average =", average)
else:
    print("No marks entered.")
```

```python
# 20 Create login system with maximum 3 attempts.
# Explanation
# Store correct username and password
# Ask user input
# If correct → Login successful
# If wrong → decrease attempts
# After 3 wrong attempts → block account

username = "admin"
password = "1234"

attempts = 3

while attempts > 0:
    user = input("Enter Username: ")
    pwd = input("Enter Password: ")

    if user == username and pwd == password:
        print("Login Successful")
        break
    else:
        attempts = attempts - 1
        print("Invalid Username or Password")
        print("Attempts Left:", attempts)

if attempts == 0:
    print("Account Blocked")
```
