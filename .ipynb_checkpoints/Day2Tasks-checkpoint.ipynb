```python
#Q4
a=int(input("Enter the first value: "))
b=int(input("Enter the 2nd value: "))
c=int(input("Enter the 3rd value: "))

if a>b and a>c:
    print("Largest :",a)
if b>a and b>c:
    print("Largest :",b)
if c>a and c>b:
    print("Largest :",c)
```
