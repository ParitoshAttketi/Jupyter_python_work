```python
# 10 Take a year and check if it is a leap year (div by 4 but not 100, or div by 400).
year=int(input("enter year"))
if (year % 4) and (year % 100 != 0) or (year % 400):
    print(f"{year} is a leap year")
else:
    print(f"{year} is not leap year")
```

```python
# 15 Take three sides of a triangle. Check if they form a valid triangle. If yes, also print if it is Equilateral, Isosceles, or Scalene.
ang1=int(input(""))
ang2=int(input(""))
ang3=int(input(""))
if (ang1 + ang2) > ang3 and (ang2 + ang3) > ang1 and (ang1 + ang3) > ang2:
    print("valid triangle")
    if ang1 == ang2 == ang3:
        print("Equilateral Triangle")
    elif ang1 == ang2 or ang2 == ang3 or ang1 == ang3:
        print("Isosceles Triangle")
    else:
        print("Scalene Triangle")
else:
    print("not a valid triangle")
```

```python
# 16 Take student marks in 5 subjects. Calculate percentage and give grade: 90+ → A+, 80-89 → A, 70-79 → B, 60-69 → C, below 60 → Fail.
subj1=int(input("enter marks"))
subj2=int(input("enter marks"))
subj3=int(input("enter marks"))
subj4=int(input("enter marks"))
subj5=int(input("enter marks"))
tmarks= subj1+subj2+subj3+subj4+subj5
print(f"Total marks : {tmarks}")
if tmarks >= 90:
    print(" A+ grade")
elif tmarks >= 80:
    print(" A grade")
elif tmarks >= 70:
    print(" B grade")
elif tmarks >= 60:
    print(" C grade")
else:
    print(" fail")
```

```python
# 20 Take three numbers. Print them in ascending order using if-elif-else (without using sort()).
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

if num1 <= num2 <= num3:
    print(num1, num2, num3)
elif num1 <= num3 <= num2:
    print(num1, num3, num2)
elif num2 <= num1 <= num3:
    print(num2, num1, num3)
elif num2 <= num3 <= num1:
    print(num2, num3, num1)
elif num3 <= num1 <= num2:
    print(num3, num1, num2)
else:
    print(num3, num2, num1)
```
