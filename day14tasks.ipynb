## Easy Level (1–10)

```python
# 1.Write a lambda function to return the square of a number.
# Example:
# Input: 5
# Output: 25


sq = lambda x : x*x
sq(2)
```

```python
# 2 Write a lambda function to return the cube of a number.
#Example:
#Input: 4
#Output: 64

cube = lambda x : x**3
cube(4)
```

```python
# 3 Write a lambda function to add two numbers.
# Example:
# Input: 10, 20
# Output: 30

add = lambda x,y : x+y
add(10,20)
```

```python
# 4 Write a lambda function to find the maximum of two numbers.
# Example:
# Input: 15, 8
# Output: 15

maximun = lambda x,y : max(x,y)
maximun(15,8)
```

```python
# 5 Write a lambda function that returns "Even" if a number is even; otherwise "Odd".
# Input: 9
# Output: Odd

even_odd = lambda x : "even" if x % 2 == 0 else "odd"
even_odd(9)
```

```python
# 6 Use map() with lambda to square every element of a list.
# Input:
# [1,2,3,4]
# Output:
# [1,4,9,16]
numbers = [1,2,3,4]
sq= list(map(lambda x: x*x, numbers))
sq
```

```python
# 7 Use filter() with lambda to keep only even numbers.
# Input:
# [1,2,3,4,5,6]
# Output:
# [2,4,6]

numbers=[1,2,3,4,5,6]
evn_odd = list(filter(lambda x: x % 2 == 0, numbers))
evn_odd
```

```python
# 8 Use filter() with lambda to keep numbers greater than 50.
# Input:
# [20,45,55,70]
# Output:
# [55,70]

numbers=[20,45,55,70]
greater= list(filter(lambda x : x >= 50, numbers))
greater
```

```python
# 9 Sort a list of numbers in ascending order using lambda.
# [9,5,1,8,2]

list1 = [9,5,1,8,2]
a= sorted(list1, key= lambda x: x)
a
```

```python
# 10 Sort a list of numbers in descending order using lambda.
# [9,5,1,8,2]

list1 = [9,5,1,8,2]
a= sorted(list1, key=lambda x: x,reverse=True)
a

```
