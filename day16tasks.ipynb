```python
# 25. Create class Employee with private __salary. Setter should not allow value below 5000.

class Employee:
    def __init__(self,salary):
        self.set_salary(salary) 
        
    def set_salary(self, salary):
        if salary >= 5000:
            self.__salary = salary
        else:
            print("Salary cannot be less than 5000.")
            self.__salary = 5000

    def get_salary(self):
        return self.__salary

emp = Employee(25000)
print("salary :",emp.get_salary())

emp1 = Employee(2000)
print("salary :",emp1.get_salary())

emp2 = Employee(4000)
print("salary :",emp2.get_salary())
```

```python
# 26. Define class Temperature with private __celsius. Add getter and setter with conversion.

class Temperature:
    def __init__(self,celsius):
        self.set_celsius(celsius)


    def set_celsius(self,celsius):
        self.__celsius = celsius

    def get_celsius(self):
        return self.__celsius

    def get_fahrenheit(self):
        return (self.__celsius * 9/5) + 32

temp = Temperature(20)

print("celsius    :", temp.get_celsius())
print("fahrenheit :", temp.get_fahrenheit())

temp.set_celsius(104)

print("Celsius:", temp.get_celsius())
print("Fahrenheit:", temp.get_fahrenheit())
```

```python
# 30. Define class Rectangle with private __length and __width. Add getter/setter and area method.

class Rectangle:
    def __init__(self, length, width):
        self.set_length(length)
        self.set_width(width)

    def set_length(self, length):
        self.__length = length

    def get_length(self):
        return self.__length

    def set_width(self, width):
        self.__width = width

    def get_width(self):
        return self.__width

    def area(self):
        return self.__length * self.__width

rect = Rectangle(10, 5)

print("Length:", rect.get_length())
print("Width :", rect.get_width())
print("Area  :", rect.area())

rect.set_length(15)
rect.set_width(8)

print("Updated Length:", rect.get_length())
print("Updated Width :", rect.get_width())
print("Updated Area  :", rect.area())
```
